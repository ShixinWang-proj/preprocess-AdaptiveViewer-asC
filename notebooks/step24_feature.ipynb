```python
import pandas as pd

# 确保时间列是 datetime 格式（如果之前已经转过，这步会自动兼容）
# 💡 注意：如果您在执行 df.rename 之前运行此代码，请将 'timestamp' 换成 '时间戳'
time_col = 'timestamp' if 'timestamp' in df.columns else '时间戳'
df[time_col] = pd.to_datetime(df[time_col])

# 获取起止时间
start_time = df[time_col].min()
end_time = df[time_col].max()

# 计算总时长
duration = end_time - start_time

print("=" * 40)
print(f"🗓️ 数据起始时间: {start_time}")
print(f"🗓️ 数据结束时间: {end_time}")
print(f"⏳ 数据总跨度:   {duration}")
print("=" * 40)
```
